# Task 1

Write function that calculates value of this graph for any given input vector `x` and set of coefficients $b_1, b_2, c_1, c_2$ packed as vector of weights `w`.
In this function also return all the $z$ values.

````python
import numpy as np

## YOU CAN DEFINE ANY FUNCTIONS YOU NEED ##
# -- placeholder start --
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)
# -- placeholder end --

def graph_value(x, w):
    y = x
    ## YOUR CODE HERE ##
    # -- placeholder start --
    x_1 = x[0]
    x_2 = x[1]

    b_1 = w[0]
    b_2 = w[1]
    c_1 = w[2]
    c_2 = w[3]

    z_1 = x_1 + b_1
    z_2 = x_2 + b_2
    z_3 = sigmoid(z_1)
    z_4 = sigmoid(z_2)
    z_5 = tanh(z_2)
    z_6 = z_5 * c_2
    z_7 = z_1 * z_4
    z_8 = z_7 * c_1
    z_9 = z_3 * z_6
    z = [z_1, z_2, z_3, z_4, z_5, z_6, z_7, z_8, z_9]
    y = z_8 + z_9
    # -- placeholder end --
    return y, z
````

````python
answer['graph_value'] = [graph_value(**input) for input in inputs]
````

# Task 2

In this graph, find all direct derivatives of all the values over $z$ or $x$ or $w$.

For example, if $y = z_8 + z_9$, you need to find only $\partial_{z_9} y$ and $\partial_{z_8} y$ for that operation.

You have to do that for all the intermediate signals.

Write the result in form of dictionary, for example, $\partial_{z_9} y$:

```
['dy_dz9'] = 1
```

````python
def graph_derivatives(x, w):
    res = {}
    ## YOUR CODE HERE ##
    # -- placeholder start --
    y, z = graph_value(x, w)

    z1, z2, z3, z4, z5, z6, z7, z8, z9 = z
    b1, b2, c1, c2 = w
    x1, x2 = x
    
    # y derivatives
    res['dy_dz9'] = 1
    res['dy_dz8'] = 1

    # z9 derivatives
    res['dz9_dz6'] = z3
    res['dz9_dz3'] = z6

    # z8 derivatives
    res['dz8_dc1'] = z7
    res['dz8_dz7'] = c1

    # z7 derivatives
    res['dz7_dz1'] = z[3]
    res['dz7_dz4'] = z[0]

    # z6 derivatives
    res['dz6_dc2'] = z[4]
    res['dz6_dz5'] = w[3]

    # z5 derivatives
    res['dz5_dz2'] = (1 - z[4]) * (1 + z[4])

    # z4 derivatives
    res['dz4_dz2'] = z[3] * (1 - z[3])

    # z3 derivatives
    res['dz3_dz1'] = z[2] * (1 - z[2])

    # z2 derivatives
    res['dz2_dx2'] = 1
    res['dz2_db2'] = 1

    # z1 derivatives
    res['dz1_dx1'] = 1
    res['dz1_db1'] = 1
    # -- placeholder end --

    return res
````

````python
answer['graph_derivatives'] = [graph_derivatives(**input) for input in inputs]
````

# Task 3

Using the values of the derivatives, calculated above:
- extract a dictionary of values that are needed to calculate $\partial_{c_1} y$
- calculate that derivative

````python
def dy_dc1(x, w):
    selected_derivs = {}
    dy_dc1 = 0

    ## YOUR CODE HERE ##
    # -- placeholder start --
    y, z = graph_value(x, w)
    derivs = graph_derivatives(x, w)

    selected_derivs['dy_dz8'] = derivs['dy_dz8']
    selected_derivs['dz8_dc1'] = derivs['dz8_dc1']

    dy_dc1 = derivs['dy_dz8'] * derivs['dz8_dc1']
    # -- placeholder end --

    return selected_derivs, dy_dc1
````

````python
answer['dy_dc1'] = [dy_dc1(**input) for input in inputs]
````

# Task 5
Using the values of the derivatives, calculated above:
- extract a dictionary of values that are needed to calculate $\partial_{b_1} y$
- calculate that derivative

````python
def dy_db1(x, w):
    selected_derivs = {}
    dy_db1 = 0

    ## YOUR CODE HERE ##
    # -- placeholder start --
    # y z8 z7 z1 b1
    # y z9 z3 z1 b1

    y, z = graph_value(x, w)
    derivs = graph_derivatives(x, w)

    selected_derivs['dy_dz8'] = derivs['dy_dz8']
    selected_derivs['dz8_dz7'] = derivs['dz8_dz7']
    selected_derivs['dz7_dz1'] = derivs['dz7_dz1']
    selected_derivs['dz1_db1'] = derivs['dz1_db1']

    selected_derivs['dy_dz9'] = derivs['dy_dz9']
    selected_derivs['dz9_dz3'] = derivs['dz9_dz3']
    selected_derivs['dz3_dz1'] = derivs['dz3_dz1']
    selected_derivs['dz1_db1'] = derivs['dz1_db1']

    dy_db1 = derivs['dy_dz9'] * derivs['dz9_dz3'] * derivs['dz3_dz1'] * derivs['dz1_db1'] + derivs['dy_dz8'] * derivs['dz8_dz7'] * derivs['dz7_dz1'] * derivs['dz1_db1']
    # -- placeholder end --

    return selected_derivs, dy_db1
````
